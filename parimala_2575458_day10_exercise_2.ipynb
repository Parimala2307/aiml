# 1.what are the most 10 common movie names of all time?

```python
 #Group the DataFrame by the 'title' column and count occurrences
title_counts = titles['title'].value_counts()

# Display the ten most common movie names
top_10_titles = title_counts.head(10)
print(top_10_titles)
```

# 2.Which three years of the 1930s saw the most films released?



```python
# Filter the DataFrame for films released in the 1930s
films_1930s = titles[(titles['year'] >= 1930) & (titles['year'] <= 1939)]

# Count the number of films released in each year and sort them in descending order
film_counts = films_1930s['year'].value_counts().sort_index(ascending=False)

# Get the top three years with the most films released
top_three_years = film_counts.head(3)

print(top_three_years)
```

# 3.Plot the number of films that have been released each decade over the history of cinema.


```python
import pandas as pd
import matplotlib.pyplot as plt

titles = pd.read_csv('cast.csv', index_col=None)

titles['decade'] = (titles['year'] // 10) * 10

decade_counts = titles['decade'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
plt.bar(decade_counts.index, decade_counts.values, width=5, align='center', tick_label=decade_counts.index)
plt.xlabel('Decade')
plt.ylabel('Number of Films Released')
plt.title('Number of Films Released Each Decade')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

# 10.Which actors or actresses appeared in the most movies in the year 1945?


```python
import pandas as pd

cast = pd.read_csv('cast.csv', index_col=None)

movies_1945 = cast[cast['year'] == 1945]

actor_appearances = movies_1945['name'].value_counts()

most_appearances_1945 = actor_appearances.idxmax()
most_appearances_count = actor_appearances.max()

print(f"The actor/actress who appeared in the most movies in 1945 is {most_appearances_1945} with {most_appearances_count} appearances.")
```

# 12. Plot how many roles Mammootty has played in each year of his career.

```python
import pandas as pd
import matplotlib.pyplot as plt

cast = pd.read_csv('cast.csv', index_col=None)

mammootty_roles = cast[cast['name'] == 'Mammootty']

roles_per_year = mammootty_roles['year'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
plt.plot(roles_per_year.index, roles_per_year.values, marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Number of Roles')
plt.title('Number of Roles Played by Mammootty Each Year')
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

# 15.Plot the n-values of the roles that Judi Dench has played over her career.
